```python
import pandas as pd
import matplotlib.pyplot as plt
colors = {0:'tab:blue',1:'tab:orange',2:'tab:green'}
```

```python
data = pd.read_pickle('data/datapickle.sec')
data
```

```python
data.info()
```

```python
# add new column time_interval into data
data['time_interval'] = data.end_time - data.start_time
```

```python
# visualize all time_interval
plt.figure(figsize=(16,8))
plt.title('time_interval')
plt.plot(data['time_interval'])
plt.xlabel('index', fontsize = 18)
plt.ylabel('time_interval', fontsize = 18)
plt.show()
```

```python
data.run.unique()
# len(data.run.unique()) : 74 trials
```

```python
# plot the time_interval for every trials
for i in data.run.unique():
    df = data[data['run'] == i]
    # visualize the time_interval
    plt.figure(figsize=(16,8))
    plt.title('time_interval' + str(i))
    plt.plot(df['time_interval'])
    plt.xlabel('index', fontsize = 18)
    plt.ylabel('time_interval', fontsize = 18)
    plt.show()
```
