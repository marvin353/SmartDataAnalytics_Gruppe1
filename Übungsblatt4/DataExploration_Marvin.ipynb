```python
def calculate_time_difference(df_local, segments, index):
    first_ts = df[df["segment"] == segments[i]]['timestamp'].iloc[0]
    last_ts = df[df["segment"] == segments[i]]['timestamp'].iloc[-1]
    diff = last_ts - first_ts
    diff_seconds = diff.seconds + diff.microseconds/1000000
    #print("Diff: " + str(diff))
    #print("DiffSec: " + str(diff_seconds))
    return diff_seconds
```

```python
# Duration distribution by segment
durations_move = []
durations_descent = []
durations_spiral = []

for run in df['run'].unique():
    df_run = df[df["run"] == run]
    segments = df_run["segment"].unique()

    #Skip run if data is wrong
    if len(segments) > 24:
        print("Skip run" + str(int(run)))
        continue # Skip run 21, 23, 40, 48

    segments_move = segments[0::3]
    segments_descend = segments[1::3]
    segments_spiral = segments[2::3]

    #Calculate time intervalls (duration in milliseconds)

    for i in range(0,8):
        diff_seconds_move = calculate_time_difference(df, segments_move, i)
        diff_seconds_descent = calculate_time_difference(df, segments_descend, i)
        diff_seconds_spiral = calculate_time_difference(df, segments_spiral, i)

        durations_move.append(diff_seconds_move)
        durations_descent.append(diff_seconds_descent)
        durations_spiral.append(diff_seconds_spiral)

        """first_ts_move = df[df["segment"] == segments_move[i]]['timestamp'].iloc[0]
        last_ts_move = df[df["segment"] == segments_move[i]]['timestamp'].iloc[-1]
        diff_move = last_ts_move - first_ts_move

        #print("First: " + str(first_ts_move))
        #print("Last: " + str(last_ts_move))
        print("Diff: " + str(diff_move))
        print("DiffSec: " + str(diff_move.seconds + diff_move.microseconds/1000000))"""

    #print(segments)
    print(segments_move)

print(durations_move)
sns.set_style('whitegrid')
#sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth":2.5})
fig, axs = plt.subplots(ncols=3, figsize=(20,6))
fig.tight_layout(pad=3.0)
axs[0].set_xlabel("Seconds to finish task: MOVE", size = 17 )
axs[1].set_xlabel("Seconds to finish task: DESCENT", size = 17 )
axs[2].set_xlabel("Seconds to finish task: SPIRAL SEARCH", size = 17 )
sns.kdeplot(np.array(durations_move), x="Seconds", bw=0.5, ax=axs[0])
sns.kdeplot(np.array(durations_descent), bw=0.5, ax=axs[1], x="Seconds")
sns.kdeplot(np.array(durations_spiral), bw=0.5, ax=axs[2], x="Seconds")

"""Diese Plots zeigen die zeitliche Verteilung (in Sekunden) beim Setzen der einzelnen Komponenten, unterteilt entsprechend der drei Segmente "Move", "Descend" und "Spiral Seach". Es ist hier zu erkennen, dass die Spiralsuche am Wenigsten und das Bewegen des Arms(Move) tendentiell am Meisten Zeit in Anspruch nimmt. Auch wenn es nicht zu dieser Aufgabe gehört, wäre für den Task "Move" sowie für den Task "Descent" eine Optimierung sinnvoll, da diese im Gegensatz zur Spiralsuche viel Zeit benötigen. """
```

```python
plt.rc('xtick', labelsize=12)    # fontsize of the tick labels
plt.rc('ytick', labelsize=12)
# take run == 0 as example
def plot_run(run: int, parameter:str):
    xl = parameter + "_x"
    yl = parameter + "_y"
    zl = parameter + "_z"
    df_temp = data[data["run"]==run]
    for i in range(int(len(df_temp["segment"].unique())/3)):
        fig = plt.figure(figsize=(15,4))
        ax = fig.add_subplot(1, 4, 1, projection='3d')
        plt.title('Inserting {} Component.'.format(i+1), fontsize=20)
        #ax.title.set_text('Inserting {} Component.'.format(i+1))
        # Plot the first column - Grey Lines
        for index, seg in enumerate(df_temp["segment"].unique()[0:i*3]):
            t = df_temp[df_temp["segment"]==seg]
            x = t[xl]
            y = t[yl]
            z = t[zl]

            ax.plot(x, y, z, color="tab:gray")
        # Plot the first column - Colored Lines
        for index, seg in enumerate(df_temp["segment"].unique()[i*3:(i+1)*3]):
            t = df_temp[df_temp["segment"]==seg]
            x = t[xl]
            y = t[yl]
            z = t[zl]

            ax.plot(x, y, z)

        # Plot Columns 1,2,3
        for index, seg in enumerate(df_temp["segment"].unique()[i*3:(i+1)*3]):
            t = df_temp[df_temp["segment"]==seg]
            x = t[xl]
            y = t[yl]
            z = t[zl]
            ax = fig.add_subplot(1,4, index%3+2, projection='3d' )
            ax.plot(x, y, z,color = colors[index])
        plt.tight_layout()

    #plt.savefig("{}.png".format(i),dpi=300,transparent=True)
plot_run(0, "pos")
```
